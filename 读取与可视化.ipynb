# 进行可视化分析

```python
%pylab inline
```

```python
df['Close'].plot(kind='line', figsize=[13,5], legend=True, title='BABA', grid=True)
```

```python
df['mvg20'] = df['Close'].rolling(window=20).mean()
```

```python
df[['Close', 'mvg20']].plot(kind='line', figsize=[13,5])
```

```python
df.ix[df.index >= '2017-07-01', 'Volume'].plot(kind='bar', title='BABA-Vol')
```

```python
df['diff'] = df['Close'] - df['Open']
df['rise'] = df['diff'] > 0
df['fall'] = df['diff'] < 0
```

```python
df[['rise', 'fall']].sum().plot(kind='pie', figsize=[5,5], counterclock=True, startangle = 90, legend=True, title='Rise Percentage')
```

```python
df['Close'].plot(kind='hist', title='Close Price Distribution')
```

```python
df['ret'] = df['Close'].pct_change(1)
```

```python
df['ret'].plot(kind= 'hist', bins = 50)
```

```python
df['Close'].plot(kind='box')
```
